```python
#Load the dataset as a Pandas data frame.
import pandas as pd
vg_sales = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')

```

```python
#Display the first ten rows of data.
vg_sales.head(10)
```

```python
#Find the dimensions (number of rows and columns) in the data frame. What do these two numbers represent in the
#context of the data?
vg_sales.shape

#this represent there are 16719 observations and 16 attributes in the data.
```

```python
#Find the top five games by critic score.
vg_sales.sort_values(by="Critic_Score", ascending=False).loc[:,['Name','Critic_Score']].head(5)
```

```python
#Find the number of video games in the data frame in each genre.
vg_sales.loc[:,['Genre']].value_counts()
```

```python
#Find the first five games in the data frame on the SNES platform.
vg_sales.loc[vg_sales['Platform']=='NES',['Name','Platform']].head()
```

```python
#Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for 
#each publisher to do this.
vg_sales.groupby(['Publisher']).agg({'Global_Sales':'sum'}).sort_values(by="Global_Sales", ascending=False).head(5)
```

```python
#Create a new column in the data frame that calculates the percentage of global sales from North America. 
#Display the first five rows of the new data frame.
vg_sales['na_global_percent'] = vg_sales.apply(lambda vg_sales: (vg_sales.NA_Sales / vg_sales.Global_Sales)*100, axis=1)
vg_sales.head(5)
```

```python
#Find the number NaN entries (missing data values) in each column.

vg_sales.isna().sum()
```

```python
#Try to calculate the median user score of all the video games. You will likely run into an error because some of the
#user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN 
#and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

median = vg_sales['User_Score'].loc[vg_sales['User_Score'] != 'tbd'].median()
vg_sales['User_Score'].fillna(value=median, inplace=True)
#vg_sales.loc[vg_sales['User_Score'].isnull(),['Name','User_Score']].head()
```
