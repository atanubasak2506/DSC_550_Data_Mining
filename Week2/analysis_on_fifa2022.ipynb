```python
import pandas as pd
pd.set_option('display.max_columns', None)
fifa = pd.read_csv('Fifa Worldcup 2022.csv')
fifa.head(10)
```

#Identify a question or question(s) that you would like to explore in your data set.

Below are my questions about the FIFA 2022 matches, that I want to explore using this data.

1. Which team belonging to which group is doing good in respect of goal count?
2. Which team are still in league till now?
3. Which Team playes more decently and which plays aggresive soccer, lets do some card analysis?

```python
#Create at least three graphs that help answer these questions. Make sure your graphs are clearly readable and are 
#labeled appropriately and professionally.

import seaborn as sns
import matplotlib.pyplot as plt

# Lets found how many groups are the there in 2022 FIFA
fifa.loc[:,['Group']].value_counts()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('WorldCup 2022 - Team by Groups and their Goal Summary')

sns.barplot(ax=axes[0, 0], data=fifa.loc[(fifa['Group'] == 'A')], x="Group", y="Goal", hue="Team", estimator=sum)
sns.barplot(ax=axes[0, 1], data=fifa.loc[(fifa['Group'] == 'B')], x="Group", y="Goal", hue="Team", estimator=sum)
sns.barplot(ax=axes[0, 2], data=fifa.loc[(fifa['Group'] == 'C')], x="Group", y="Goal", hue="Team", estimator=sum)
sns.barplot(ax=axes[1, 0], data=fifa.loc[(fifa['Group'] == 'D')], x="Group", y="Goal", hue="Team", estimator=sum)
sns.barplot(ax=axes[1, 1], data=fifa.loc[(fifa['Group'] == 'E')], x="Group", y="Goal", hue="Team", estimator=sum)
sns.barplot(ax=axes[1, 2], data=fifa.loc[(fifa['Group'] == 'F')], x="Group", y="Goal", hue="Team", estimator=sum)
plt.show()
```

I was trying to see which team belonging to which group and doing highest amount of goals. In the above plot we can say Netherland is having the highest goal from Group A, England from Group B, Argentina from Group C, France from Group D, Spain from group E and Croatia from Group F has highest number of Goals.

```python
# Let find the red cards and yellow cards for by the team, to understand which team is playing more aggrasively.
fig, axes = plt.subplots(1, 2, figsize=(18, 10))
fig.suptitle('Red Card and Yellow Card Summary by Team')

sns.barplot(ax=axes[0], data=fifa, x="Red Cards", y="Team",  estimator=sum, color='red')
sns.barplot(ax=axes[1], data=fifa, x="Yellow Cards", y="Team",  estimator=sum, color='yellow')
plt.show()
```

I have done the analysis on the yellow and red cards in the soccer match for worldcup 2022 by different team. 
When I look at the yellow cards England is the team which have not received any yellow or red card seems like they play obeying the scoccer rules, 
then Argentica, France, Spain does not play aggresively and follow soccer rules, looking at the yellow card summary team like Saudi Arabia and Serbia have
very high count of yellow cards, which seems like team doesn't follow the soccer rules properly, if we look at the red card two team player from Wales and Cameron got red cards.
